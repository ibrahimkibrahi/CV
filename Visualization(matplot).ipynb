```python
import numpy as np
import matplotlib.pyplot as plt
```

## Life cycle of a figue

In other to create clean figues, the following life cycle must be followed

1. plt.figue(figsize())
2. plt.plot()
3. Extras (title, axes, legeds)
4. plt.show()

```python
x = np.linspace(0, 2, 10)
plt.figure(figsize=(12,16)) # size of the figue
plt.plot(x,x**2,label="First Graph in matplot")
plt.plot(x,x**3, label = "second graph in matplot")

#Extra information
plt.suptitle("figue 1") # Title
plt.title("Yea")
plt.xlabel("axe x") # axes x
plt.ylabel("axe y") # axes
plt.legend() # legend

plt.savefig("figue.png")
plt.show()
```

### Subplots

subplots are another element to add to create several graphics on the same figure

```python
x = np.linspace(0,2,10)

y = x**2
plt.figure(figsize=(12,6))
plt.subplot(2,2,1)# we are going to have 2 rows, 2 columns and this is the first panel
plt.plot(x ,y, lw =4, c="red")
plt.subplot(2,2,2)
plt.plot(x,y, lw =4, c="blue")
plt.subplot(2,2,3)
plt.plot(x ,y, lw =4, c="red")
plt.subplot(2,2,4)
plt.plot(x,y, lw =4, c="blue")
```

## Using the Oriented Object Method

```python
fig, ax = plt.subplots(2,1, sharex=True)# shares the same axis for subplots
ax[0].plot(x,y)
ax[1].plot(x,np.sin(x))
plt.show()
```

```python
x= np.arange(-50,50.1,0.1)
x2 = np.arange(0,50.1,0.1)
y1 = np.array([(1/2)*i + 1 for i in x])
y2 = x**2
y3 = (2*(x**2) + (2**x))
y4 = np.sin(x2**(1/2))

my_figure = plt.subplots(nrows=2, ncols=2, figsize=(14,6))
fig, ((ax1,ax2),(ax3,ax4)) = my_figure

ax1.plot(y1, c="crimson")
ax2.plot(y2, c="black")
ax3.plot(y3, c="blue")
ax4.plot(y4, c="purple")
```

```python
dataset = {f"experience{i}": np.random.randn(100) for i in range (4)}
def graphique(data):
  n = len(data)
  plt.figure(figsize=(12,10))
  for k, i in zip (data.keys(), range(1, n+1)):
    plt.subplot(n,1,i)
    plt.plot(data[k])
    plt.title(k)
  plt.show()
```

```python
graphique(dataset)
```

## Scatter

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
x=iris.data
y=iris.target
```

```python
plt.scatter(x[:,0], x[:,i], c=y)
```

```python
n = x.shape[0]
plt.figure(figsize=(12,8))
for i in range(n):
  plt.subplot(n//2, n//2, i+1)
  plt.scatter(x[:,0], x[:,i], c=y)
  plt.xlabel("0")
  plt.ylabel(i)
  plt.colorbar(ticks=(np.unique(y)))
```

```python
f= lambda x, y: np.sin(x) + np.sin(x+y)
x= np.linspace(0,5,50)
y = np.linspace(0,5,50)
x, y = np.meshgrid(x,y)
z = f(x,y)

ax = plt.axes(projection="3d")
ax.plot_surface(x,y,z,cmap="Blues")
plt.show()
```
